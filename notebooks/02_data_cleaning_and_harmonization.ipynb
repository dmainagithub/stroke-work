#### Loading data

```python
# Use DATA_DIR from the config
agincourt_file = DATA_DIR / "df_processed_merged_final_agincourt_site.csv"
nairobi_file = DATA_DIR / "df_processed_merged_final_nairobi_site.csv"

# Load both CSVs (Agincourt and Nairobi)
df_agincourt = pd.read_csv(agincourt_file, low_memory=False)
df_nairobi = pd.read_csv(nairobi_file, low_memory=False)

# Add site identifier before concatenating
df_agincourt['site'] = 'agincourt'
df_nairobi['site'] = 'nairobi'

# Concatenate vertically (append rows)
df_combined = pd.concat([df_agincourt, df_nairobi], 
                       ignore_index=True,  # Reset index
                       sort=False)  # Keep column order

print(f"Agincourt shape: {df_agincourt.shape}")
print(f"Nairobi shape: {df_nairobi.shape}")
print(f"Combined shape: {df_combined.shape}")
df = df_combined.copy()
```
